```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchvision.datasets import FashionMNIST
```

```python
# 실행마다 동일한 결과를 얻기 위해 파이토치에 랜덤 시드를 지정하고 GPU 연산을 결정적으로 만듭니다.
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True
```

```python
# 모델링
model = nn.Sequential()
model.add_module('conv1', nn.Conv2d(1, 32, kernel_size=3, padding='same')) # (3,3)크기 입력 1, 출력 32로 입출력 같은 크기로 패딩 >> (32, 28, 28)
model.add_module('relu1', nn.ReLU()) # relu 활성화 함수
model.add_module('pool1', nn.MaxPool2d(2)) # (2,2)크기 최대풀링 >> (32, 14, 14)
model.add_module('conv2', nn.Conv2d(32, 64, kernel_size=3, padding='same')) # (3,3)크기 입력 32, 출력 64로 입출력 같은 크기로 패딩 >> (64, 14, 14)
model.add_module('relu2', nn.ReLU()) # relu 활성화 함수
model.add_module('pool2', nn.MaxPool2d(2)) # (2,2)크기 최대풀링 >> (64, 7, 7)
model.add_module('flatten', nn.Flatten()) # flatten 활성화 함수
model.add_module('dense1', nn.Linear(3136, 100)) # 입력 3136(64 * 7 * 7), 출력 100 linear 모델
model.add_module('relu3', nn.ReLU()) # relu 활성화 함수
model.add_module('dropout', nn.Dropout(0.3)) # 30% 탈락 학습
model.add_module('dense2', nn.Linear(100, 10)) # 입력 100, 출력 10 linear 모델
```

```python
# 모델 가중치 로드
model.load_state_dict(torch.load('best_cnn_model.pt', weights_only=True))

# 각 추가된 층 정의
layers = [layer for layer in model.children()]

# 추가된 1번 층 확인
print(layers[0])

# 추가된 1번 층 확인
print(model[0])
```

```python
# 각 추가된 층 정의 확인
for name, layer in model.named_children():
    print(f"{name:10s}", layer)
```

```python
# conv1 층 정의 확인
print(model.conv1)
```

```python
# conv1 층 가중치 평균 & 표준편차
conv_weights = model.conv1.weight.data
print(conv_weights.mean(), conv_weights.std())
```

```python
# conv1 층 가중치 그래프
plt.hist(conv_weights.reshape(-1, 1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
# conv1 가중치 tensor 확인
print(conv_weights.shape)
```

```python
# conv1 각 가중치 시각화
fig, axs = plt.subplots(2, 16, figsize=(15,2))
for i in range(2):
    for j in range(16):
        axs[i, j].imshow(conv_weights[i*16 + j,0,:,:], vmin=-0.5, vmax=0.5)
        axs[i, j].axis('off')
plt.show()
```

```python
# 훈련 데이터 로드
fm_train = FashionMNIST(root='.', train=True, download=True)
train_input = fm_train.data

# 1번 샘플 그래프
plt.imshow(train_input[0], cmap='gray_r')
plt.show()
```

```python
# 1번 샘플 스케일링
ankle_boot = train_input[0:1].reshape(1, 1, 28, 28) / 255.0

# 1번 샘플 conv1 & relu 함수 적용
model.eval()
with torch.no_grad():
    feature_maps = model.conv1(ankle_boot)
    feature_maps = model.relu1(feature_maps)

print(feature_maps.shape)
```

```python
# 1번 샘플 각 가중치 시각화
fig, axs = plt.subplots(4, 8, figsize=(15,8))
for i in range(4):
    for j in range(8):
        axs[i, j].imshow(feature_maps[0,i*8 + j,:,:])
        axs[i, j].axis('off')
plt.show()
```

```python
# 1번 샘플 conv1 >> relu1 >> pool1 >> conv2 >> relu2 적용
model.eval()
with torch.no_grad():
    feature_maps = model.conv1(ankle_boot)
    feature_maps = model.relu1(feature_maps)
    feature_maps = model.pool1(feature_maps)
    feature_maps = model.conv2(feature_maps)
    feature_maps = model.relu2(feature_maps)
```

```python
# 각 모델 정의 확인
model.eval()
x = ankle_boot
with torch.no_grad():
    for name, layer in model.named_children():
        x = layer(x)
        if name == 'relu2':
            break
feature_maps = x
```

```python
# 1번 샘플 특성 맵 시각화
fig, axs = plt.subplots(8, 8, figsize=(12,12))
for i in range(8):
    for j in range(8):
        axs[i, j].imshow(feature_maps[0,i*8 + j,:,:])
        axs[i, j].axis('off')
plt.show()
```
